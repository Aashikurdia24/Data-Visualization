## SEABORN

```python
# importing datasets , lib 
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('tips')
df.head(5)
```

```python
# available datasets in sns
print(sns.get_dataset_names())
```

```python
df.info()
```

```python
# Basic scatter plot
sns.set_style("white")
sns.scatterplot(x = "total_bill" , y = "tip" ,data = df)
plt.title(" Total bill by Tip ")
plt.show()
```

Observatio : as total bill increase tip also increase 

```python
# using hue
sns.scatterplot(x = "total_bill" , y = "tip" ,data = df , hue = "smoker" ,hue_order = ['Yes','No'])
plt.show()
```

Observation : non - smoker have comparatively more tip as per they total bill

```python
# using hue order color
hue_colors = {'Yes' : 'magenta','No' :'crimson'}
sns.scatterplot(x = "total_bill" , y = "tip" ,data = df , hue = "smoker" ,hue_order = ['Yes','No'] , palette = hue_colors )
plt.show()
```

## Relational plots and subplots

```python
# using col --> side by side and 
# row --> up and down
sns.relplot(x = "total_bill" , y = "tip" ,data = df , col = "smoker" , kind = "scatter", color = "crimson" )
plt.show()
```

```python
sns.relplot(x = "total_bill" , y = "tip" , data = df , row = "smoker" , kind = "scatter" , color = "crimson")
plt.show()
```

```python
# using col and row together 
# col -> smoker 
# row -> time or day 

sns.relplot(x = "total_bill" , y = "tip" , col = "smoker" , row = "time" , data = df , color = "crimson")
plt.show()
```

Observation : In lunch time less smokers visit as compared to dinner time 
              In lunch time less non - smokers visit as compared to dinner time 
              people visit more in dinner time 

```python
sns.relplot(x = "total_bill" , y = "tip" , col = "smoker" , row = "day" , data = df , color = "crimson")
plt.show()
```

```python
sns.relplot(x = "total_bill" , y = "tip" , col = "day" ,hue = "smoker" , data = df , color = "blue" , col_wrap = 4,size = "smoker")
plt.show()
```

Subgroups with point size and hue 

```python
sns.relplot(x="total_bill" , y = "tip" , data = df , kind = "scatter" , hue = "size" , size = "size")
plt.show()
```

```python
sns.relplot(x="total_bill" , y = "tip" , data = df , kind = "scatter" , hue = "size" , size = "size" , alpha = 0.8)
plt.show()
```

Line plot

```python
sns.relplot(x = "total_bill" , y = "tip" , kind = "line" , data = df , style = "smoker" , marker = True , dashes = False , hue = "smoker")
plt.show()
```

## Categorical plots

count plot , bar plot , box plot , point plot 

```python
#count plot
cat_order = ['Yes','No']
sns.catplot(x = "smoker" , data = df , kind = "count" , order = cat_order)
plt.show()
```

```python
# bar plot --> display the mean of numerical variable as per category
sns.catplot(x = "day" , y = "tip" , data = df , kind = "bar" , errorbar = 'sd')
plt.show()
# errorbar or ci = none -- > will hide the confidence interval 
# this plot display --> standard deviation and many more 
```

```python
# box plot 
sns.catplot(x = 'time' , y = 'total_bill' , data = df , kind = 'box' , order = ['Dinner','Lunch'])
plt.show()
```

```python
# box plot omitting outliers using sym
sns.catplot(x = 'time' , y = 'total_bill' , data = df , kind = 'box' , order = ['Dinner','Lunch'] , showfliers = False )
plt.show()
```

```python
# point plot 
sns.catplot(x = 'smoker' , y = 'total_bill' , data = df , kind = 'point' , join = False , capsize = 0.2, linestyle = 'none')
plt.show()
```
