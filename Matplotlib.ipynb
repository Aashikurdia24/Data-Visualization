## Matplotlib 

```python
# plotting multiple line graphs
# import matplotlib , numpy

import matplotlib.pyplot as plt
import numpy as np

# growth-rates of linear , quadratic , cubic , logarithmic , log linear , exponential 
# using linspaces() method 

xvalue = np.linspace(1,5,100)
y_lin = np.linspace(1,5,100)
y_quad = xvalue * xvalue 
y_cub = xvalue * xvalue * xvalue
y_logn = np.log(xvalue)
y_nlogn = xvalue * np.log(xvalue)
y_exp = 2 ** xvalue 

#plotting 
plt.plot(xvalue , y_lin , label = "Linear")
plt.plot(xvalue , y_quad , label = "Quadratic")
plt.plot(xvalue , y_cub , label = "Cubic")
plt.plot(xvalue , y_logn , label = "Logarithmic")
plt.plot(xvalue , y_nlogn , label = "Log linear")
plt.plot(xvalue , y_exp , label = "Exponential")

plt.legend()
plt.title("Rate of growth")
plt.xlabel("x values ")
plt.ylabel("y values ")
plt.grid()
plt.show()
```

```python
#plotting twin axes or dual axes or secondary axes 
# importing matplotlib , numpy 
# using twinx() 

import matplotlib.pyplot as plt
import numpy as np

x_value = np.linspace(1,5,100)
x_sqrt = x_value ** 0.5

# figure -> top-level container that holds all the elements 
# subplots() -> return axes object at a given grid position

(fig,ax1) = plt.subplots()
# it plot x value against x value
ax1.plot(x_value,x_value)

# dual axes for square root 
ax2 = ax1.twinx()
ax2.plot(x_value,x_sqrt)
```

```python
# plot dual axes along option
# like  labels , color , legend , axes color , title

import matplotlib.pyplot as plt
import numpy as np

x_value = np.linspace(1,5,100)
x_sqrt = x_value ** 0.5
(fig,ax1) = plt.subplots()

color1 = "crimson"
ax1.plot(x_value,x_value,color = color1)
ax1.set_ylabel("Linear", color = color1)
ax1.tick_params(axis="y",labelcolor=color1)

color2 = "blue"
ax2 = ax1.twinx()
ax2.plot(x_value,x_sqrt,color = color2)
ax2.set_ylabel("Square Root",color= color2) # label the axes 
ax2.tick_params(axis="y",labelcolor = color2)

plt.title(" Dual axes or Secondary axes ",color="pink")
plt.xlabel("scale")
```

```python
# plotting box or whisker plot
# return statistic values
# like Qualtiles , Median , IQR , outliers

#importing matplotlib , numpy
import matplotlib.pyplot as plt
import numpy as np

d1 = np.random.normal(100,10,200)
d2 = np.random.normal(90,20,200)
d3 = np.random.normal(80,30,200)

fig = plt.figure(figsize =(3, 3)) 
ax = fig.add_axes([0, 0, 1, 1])
data = [d1,d2,d3]
ax.boxplot(data, labels = ['D1','D2','D3'], patch_artist = True, showmeans = True)
plt.title("Random value")
plt.show()
```

```python
# plotting violin plot 
# importing matplotlib , numpy

import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(1)
data = np.random.randn(100)

# Create a violin plot
plt.figure()
plt.violinplot(data, showmedians=True)

# Set plot labels and title
plt.xlabel('Variable')
plt.ylabel('Value')
plt.title('Univariate Violin Plot')

# Show the plot
plt.show()
```

```python
# bar plot 
# importing matplotlib , pandas

import matplotlib.pyplot as plt
import pandas as pd

file = pd.read_excel("C:\\Users\\Admin\\Downloads\\sales.xlsx")
file1 = file.fillna(0)
file1.head(2)

month = ['April','May','June','July','Aug','Sept','Oct','Nov','Dec','Jan','Feb','Mar']
sale = []

for mon in month:
    ele = file1.loc[0: , mon].sum(axis=0).astype(int)
    sale.append(ele)

plt.figure(figsize=(8,5))
plt.bar(month,sale,color="crimson")
plt.xlabel("Monthly Sale (21-22)")
plt.ylabel("Sales figure in cr ")
plt.title("Sales 21-22 ")
```

```python
# scatter plot
# importing matplotlib , pandas

import matplotlib.pyplot as plt
import pandas as pd

file = pd.read_excel("C:\\Users\\Admin\\Downloads\\sales.xlsx")
file1 = file.fillna(0)
file1.head(2)

month = ['April','May','June','July','Aug','Sept','Oct','Nov','Dec','Jan','Feb','Mar']
sale = []

for mon in month:
    ele = file1.loc[0: , mon].sum(axis=0).astype(int)
    sale.append(ele)
         
plt.figure(figsize=(8,5))
plt.scatter(month,sale,color="crimson",marker="D",s=30)
plt.plot(month,sale,color="blue",linestyle="--",linewidth=1)
plt.grid()
plt.xlabel("Months (21-22)")
plt.ylabel("Sales figure in cr")
plt.title("Yearly sales 21-22")
```

```python
x = [10,15,20,25,30,35,40]
y = [2,4,6,8,10,12,14]

fig , ax = plt.subplots()
ax.plot(x,y,marker='o',color='r',linestyle='--')
ax.set_ylabel('y as per x ')
ax.set_xlabel('x values')
ax.annotate("local max",xy=(25,8),xytext=(23,12),arrowprops=dict(facecolor='black',shrink = 0.05),rotation = 0)

plt.title("Annotation")
plt.grid()
plt.show()
```
